Assigning a Unique ID to each Source-Destination Trip. This is done to partition the table by sourceid, destination id and time period. Doing this step ensures that the average travel time for each Source-Destination Trip in each time period chosen is correct.

An example of this is showcased below: Andheri West to Nariman Point

```python
import pandas as pd
```

```python
df1=pd.read_csv(
    "/Users/Harsh/Desktop/Uber_Movement/mumbai/Nariman_Point/mumbai_Andheri_West_to_Nariman_Point.csv",
    delimiter = ',')
```

```python
df1
```

```python
import numpy as np
df1["unique_id"] = 0
```

```python
i = 0
group_of_twelve = 0
while (i<len(df1)): 
    df1.unique_id[i] = group_of_twelve*12 + df1.time_period[i]
    if ((df1.time_period[i]==12) & (df1.row_num[i]==df1.count_of_rows[i])):
        group_of_twelve = group_of_twelve+1
    i=i+1
```
